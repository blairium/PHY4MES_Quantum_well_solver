```python
#Required Packages
import numpy as np #Provides higher level mathematical functions
import matplotlib.pyplot as plt #Provides plotting
from scipy.constants import m_e, hbar, e, epsilon_0, h #Importing constants

#Optional Figure Parameters
fig_width_pt = 400  # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27               # Convert pt to inches
golden_mean = (np.sqrt(5)-1.0)/2.0         # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches
fig_height =fig_width*golden_mean       # height in inches
fig_size = [fig_width,fig_height]       
SMALL_SIZE = 4
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=10)          # controls default text sizes
plt.rc('axes', titlesize=10)     # fontsize of the axes title
plt.rc('axes', labelsize=10)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=8)    # fontsize of the tick labels
plt.rc('ytick', labelsize=8)    # fontsize of the tick labels
plt.rc('legend', fontsize=8)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = fig_size
plt.rcParams['figure.dpi'] = 200 #Controls figure size in jupyter
```

```python
def energy_level(n, W):
    '''
    Calculates the energy values for a given integer n and well width W (in m)
    '''
    if type(n) != int: #checks if n is an integer
        print('n must be an integer!!!!')
    else:
    
        return (h**2 * n**2)/(e*8*m_e*W**2)

def well_width(n, eV, m_r): 
    '''Calculates the well width such that there are n states for a given energy level (eV). 
    m_r gives the relative electron mass
    Used for Part D'''
    return np.sqrt((h**2 * n**2)/(eV*e*8*m_r*m_e))
```

```python
energy_level(3,12e-9)
```

```python
n = np.arange(1,11)
```

```python
En_3nm = energy_level(n,3e-9)
En_6nm = energy_level(n,6e-9)
En_9nm = energy_level(n,9e-9)
```

```python
with plt.style.context(['science', 'scatter']):
    fig, ax = plt.subplots()
    ax.plot(n,En_3nm, label='3nm')
    ax.plot(n,En_6nm, label='6nm')
    ax.plot(n,En_9nm, label='9nm')
    ax.legend(title='Well width', loc=2)
    xlbl = r"$n$"
    ylbl = r"Energy eigenvalue $(eV)$"
    ax.set_xlabel(xlbl)
    ax.set_ylabel(ylbl)

    #fig.savefig('figures/q1_fig.png', dpi=1000)
    plt.show()
```
